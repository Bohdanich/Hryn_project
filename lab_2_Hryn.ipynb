### Відповіді на контрольні запитання

1. У Python існує ієрархія операторів, яка визначає порядок виконання різних математичних та логічних операцій. Нижче наведена ієрархія операторів у порядку пріоритету від найвищого до найнижчого:

** - Експонентація <br>
+x, -x - Унарні плюс та мінус <br>
*, /, %, // - Множення, ділення, остача від ділення та цілочисельне ділення <br>
+, - - Додавання та віднімання <br>
==, !=, >, >=, <, <=, is, is not, in, not in - Порівняння та перевірка ідентичності <br>
not - Логічне заперечення <br>
and - Логічне "І" <br>
or - Логічне "АБО" <br>
if-else - Умовний оператор <br>
У разі однакового пріоритету виконання операцій, порядок виконання може бути визначений асоціативністю - зліва направо або справа наліво. Наприклад, операції додавання та віднімання мають однаковий пріоритет та асоціативність зліва направо, тому вони виконуються в порядку зустрічі зліва направо.
